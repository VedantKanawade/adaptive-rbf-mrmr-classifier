```python
import sys
sys.path.append('../src')

import numpy as np
import scipy.io
from rbf_network import evaluate_sigma

X = scipy.io.loadmat('../data/data_train.mat')['data_train']
y = scipy.io.loadmat('../data/label_train.mat')['label_train'].flatten()

print(f"Data loaded: X {X.shape}, y {y.shape}")
```

```python
test_sigma = 5.0

print("Testing evaluation pipeline...")
print("=" * 50)

cv_acc, H, selected = evaluate_sigma(test_sigma, X, y, verbose=True)

print("=" * 50)
print(f"RESULT:")
print(f"  CV Accuracy: {cv_acc:.4f}")
print(f"  H neurons  : {H}")
print(f"  Selected   : {selected[:10]}... (showing first 10)")
```
